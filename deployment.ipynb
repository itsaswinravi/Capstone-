```python
for col in encoders:
    le = encoders[col]

    # handle unseen values
    if new_data[col][0] not in le.classes_:
        new_data[col] = 0
    else:
        new_data[col] = le.transform(new_data[col])
```

```python
new_data = new_data.reindex(columns=model.feature_names_in_, fill_value=0)
```

```python
pred = model.predict(new_data)
print(pred)
```
